```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import seaborn as sns
import nltk
import warnings
```

# Importing Datasets

```python
books = pd.read_csv('Books.csv', engine='python')
ratings = pd.read_csv('Ratings.csv')
users = pd.read_csv('Users.csv')
display(books.head())
display(ratings.head())
display(users.head())
```

Dropping Image URL columns from books dataset:

```python
books.drop(columns = ['Image-URL-S'], inplace = True)
books.drop(columns = ['Image-URL-M'], inplace = True)
books.drop(columns = ['Image-URL-L'], inplace = True)
```

```python
print(books.shape)
print(ratings.shape)
print(users.shape)
```
